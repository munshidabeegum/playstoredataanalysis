```python
import pandas as pd
import numpy as np
def convert_size(size):
    if size =='varies with device':
        return float(0)
    elif 'M' in size:
        return float(size.replace('M',''))
    else:
        return np.nan
df['size_KB']=df['Size'].apply(convert_size)    
```
